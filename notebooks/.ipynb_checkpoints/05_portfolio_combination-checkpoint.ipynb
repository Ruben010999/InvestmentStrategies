# Portfolio Combination: Equal-Weight and Risk Parity

**Goal.** Combine multiple strategies (e.g., Buy & Hold, Momentum, Vol-Targeted variants) into a single multi-strategy portfolio.  
We will compare a simple equal-weight approach versus a risk-parity allocation computed from rolling covariance of strategy returns.


# Step 1 — Imports & Config

```python

import sqlite3
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["axes.grid"] = True

# --- Database & universe (adjust as needed) ---
DB_PATH = Path(r"C:\Users\Ruben\Desktop\Projects\InvestmentStrategies\market.db")
TABLE     = "prices"                      # expects columns: date, symbol, close
TICKERS   = ["VTI","SPY","QQQ","AAPL","MSFT"]

# --- Backtest window ---
START     = "2015-01-01"
END       = "2025-01-01"

# --- Conventions ---
TRADING_DAYS = 252
REBAL        = "ME"                       # 'D' for daily; 'M' or 'ME' for month-end
COST_BP      = 10                         # per unit turnover at the asset level
COST_BP_STR  = 0                          # per unit turnover at the *strategy* layer (keep 0 by default)

# --- Vol targeting defaults (to reuse from previous notebook) ---
VTARGET   = 0.12
EST_W     = 60
LEV_MAX   = 1.5
```

## Step 2 — Load Prices

**Short.** Query the wide price panel (rows = dates, columns = symbols, values = adjusted close).  
We use parameter binding and pivot to a wide matrix suitable for vectorized operations.


```python

placeholders = ",".join([f":s{i}" for i,_ in enumerate(TICKERS)])
q = f"""
SELECT date, symbol, close
FROM {TABLE}
WHERE symbol IN ({placeholders})
  AND date BETWEEN :start AND :end
ORDER BY date ASC
"""
params = {**{f"s{i}": t for i, t in enumerate(TICKERS)},
          "start": START, "end": END}

with sqlite3.connect(DB_PATH) as con:
    df = pd.read_sql_query(q, con, params=params, parse_dates=["date"])

prices = (df.pivot(index="date", columns="symbol", values="close")
            .sort_index()
            .dropna(how="all"))

# Basic hygiene: forward-fill, drop any remaining NaNs, deduplicate index
prices = prices.loc[prices.index.drop_duplicates()].ffill().dropna(how="any")

# Quick sanity
print("Prices shape:", prices.shape)
print("Symbols:", list(prices.columns))
print(prices.index.min().date(), "→", prices.index.max().date())
prices.head()
```

## Step 3 — Utilities (Weights, Rebalancing, Costs, Metrics)

**Short.** Reuse helper functions to normalize weights, apply rebalancing, compute portfolio returns with costs, and summarize performance.

**Portfolio return with costs.**  
Let daily asset returns be $r_{i,t}$ and weights $w_{i,t}$. The net portfolio return is
$$
r_t^{\text{net}} \;=\; \sum_i w_{i,t}\,r_{i,t} \;-\; \tfrac{1}{2}\left(\sum_i |w_{i,t}-w_{i,t-1}|\right)\,c,
$$
where $c$ is the per-unit transaction cost (in decimal, e.g., `cost_bp × 10^{-4}`).


```python

def normalize_rows(df: pd.DataFrame) -> pd.DataFrame:
    s = df.sum(axis=1).replace(0, np.nan)
    return df.div(s, axis=0).fillna(0.0)

def apply_rebalancing(weights: pd.DataFrame, freq: str = "D") -> pd.DataFrame:
    f = freq.upper()
    if f in ("M", "ME"):
        monthly = weights.resample("ME").last()
        return monthly.reindex(weights.index).ffill().fillna(0.0)
    elif f == "D":
        return weights
    else:
        raise ValueError("freq must be 'D', 'M', or 'ME'")

def portfolio_returns(returns: pd.DataFrame,
                      weights: pd.DataFrame,
                      cost_bp: int = 0) -> pd.Series:
        returns = returns.reindex(weights.index).fillna(0.0)
        port = (returns * weights).sum(axis=1)
        if cost_bp and cost_bp > 0:
            # one-way turnover proxy: 0.5 * L1 change in weights
            dw = (weights - weights.shift(1)).abs().sum(axis=1) * 0.5
            port = port - dw * (cost_bp * 1e-4)
        return port

def cagr_from_nav(nav: pd.Series, trading_days: int = TRADING_DAYS) -> float:
    nav = nav.dropna()
    if nav.empty:
        return 0.0
    years = len(nav) / trading_days
    return float((nav.iloc[-1] / nav.iloc[0])**(1/years) - 1)

def sharpe(returns: pd.Series, trading_days: int = TRADING_DAYS) -> float:
    r = returns.dropna()
    if r.empty:
        return 0.0
    mu = r.mean() * trading_days
    sd = r.std(ddof=0) * np.sqrt(trading_days)
    return float(mu / sd) if sd > 0 else 0.0

def max_drawdown(nav: pd.Series) -> float:
    roll_max = nav.cummax()
    dd = nav / roll_max - 1.0
    return float(dd.min())

def report_from_returns(r: pd.Series, name: str) -> pd.DataFrame:
    nav = (1 + r).cumprod()
    out = pd.DataFrame({
        "Strategy": [name],
        "CAGR": [cagr_from_nav(nav)],
        "Volatility": [r.std(ddof=0) * np.sqrt(TRADING_DAYS)],
        "Sharpe(≈)": [sharpe(r)],
        "MaxDrawdown": [max_drawdown(nav)],
    }).set_index("Strategy")
    return out.round(6)
```

## Step 4 — Baseline Strategies (Returns & Weights)

**Short.** Implement baseline strategies as `(returns, weights)` pairs to make them composable later:
- **B&H Equal**: equal-weight across symbols with the chosen rebalancing,
- **12M-1 Momentum**: rank by 12-month return (skip last month) and allocate equally to positive scorers.

Weights are shifted by one day to avoid look-ahead.


```python

def bh_equal_rw(prices: pd.DataFrame, rebal: str = "ME") -> tuple[pd.Series, pd.DataFrame]:
    rets = prices.pct_change()
    w = pd.DataFrame(1.0, index=prices.index, columns=prices.columns)
    w = normalize_rows(w)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)  # execute next day
    r = (rets * w).sum(axis=1)  # raw returns (no costs here)
    return r, w

def momentum_12m_rw(prices: pd.DataFrame,
                    lookback: int = TRADING_DAYS,
                    skip_last_month: bool = True,
                    top_k: int | None = None,
                    rebal: str = "ME") -> tuple[pd.Series, pd.DataFrame]:
    rets = prices.pct_change()
    shift_curr = 21 if skip_last_month else 0
    mom = prices.shift(shift_curr) / prices.shift(lookback) - 1.0
    mom_m = mom.resample("ME").last()

    if top_k is not None and top_k > 0:
        ranks = mom_m.rank(axis=1, method="first", ascending=False)
        sel = (ranks <= top_k).astype(float)
    else:
        sel = (mom_m > 0).astype(float)

    w_m = normalize_rows(sel)
    w = w_m.reindex(prices.index).ffill().fillna(0.0)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)
    r = (rets * w).sum(axis=1)
    return r, w
```

## Step 5 — Volatility Targeting Wrapper

**Short.** Wrap raw strategy `(returns, weights)` with an ex-ante volatility scaler targeting annualized $σ^\*$.  
Scaling uses a rolling estimator with a one-day execution lag to avoid look-ahead.


```python

def rolling_annualized_vol(returns: pd.Series, window: int = 60, trading_days: int = TRADING_DAYS) -> pd.Series:
    vol = returns.rolling(window, min_periods=window).std(ddof=0)
    return vol * np.sqrt(trading_days)

def apply_vol_targeting(raw_returns: pd.Series,
                        raw_weights: pd.DataFrame,
                        target_vol: float = VTARGET,
                        est_window: int = EST_W,
                        max_leverage: float = LEV_MAX) -> tuple[pd.Series, pd.DataFrame]:
    est_vol = rolling_annualized_vol(raw_returns, window=est_window)
    scale = (target_vol / est_vol).clip(lower=0.0, upper=max_leverage).shift(1).fillna(0.0)
    vt_weights = (raw_weights.mul(scale, axis=0)).fillna(0.0)
    vt_returns = (raw_returns * scale).fillna(0.0)
    return vt_returns, vt_weights
```

## Step 6 — Construct the Strategy Set to Combine

**Short.** Build a panel of daily strategy returns that we will combine at the *strategy layer*:
- raw and volatility-targeted variants, all aligned on the same date index.


```python

rets_assets = prices.pct_change()

# 1) Raw baselines
bh_raw,  bh_w  = bh_equal_rw(prices, rebal=REBAL)
mom_raw, mom_w = momentum_12m_rw(prices, lookback=TRADING_DAYS, skip_last_month=True, top_k=None, rebal=REBAL)

# 2) Vol-targeted variants
bh_vt_r,  bh_vt_w  = apply_vol_targeting(bh_raw,  bh_w,  target_vol=VTARGET, est_window=EST_W, max_leverage=LEV_MAX)
mom_vt_r, mom_vt_w = apply_vol_targeting(mom_raw, mom_w, target_vol=VTARGET, est_window=EST_W, max_leverage=LEV_MAX)

# 3) Strategy-layer net returns (apply costs for strategy switching if desired)
def strategy_layer_net(returns: pd.DataFrame, strat_returns: pd.Series,
                       prev_strat: pd.Series | None = None, cost_bp: int = 0) -> pd.Series:
    """
    Here we assume 'strat_returns' already embeds asset-level trading costs if any.
    If you want to model switching costs between strategies, you'd need strategy weights (not used here).
    We keep COST_BP_STR = 0 by default at the strategy layer.
    """
    if cost_bp and prev_strat is not None:
        # Example placeholder for strategy switching cost, if strategy weights were used.
        pass
    return strat_returns

strategies = pd.DataFrame({
    "B&H (raw)":           strategy_layer_net(rets_assets, bh_raw,  cost_bp=COST_BP_STR),
    "B&H (vol-targeted)":  strategy_layer_net(rets_assets, bh_vt_r, cost_bp=COST_BP_STR),
    "12M-1 (raw)":         strategy_layer_net(rets_assets, mom_raw, cost_bp=COST_BP_STR),
    "12M-1 (vol-targeted)":strategy_layer_net(rets_assets, mom_vt_r, cost_bp=COST_BP_STR),
}).dropna(how="any")

print("Strategies return panel:", strategies.shape)
strategies.tail()
```

## Step 7 — Equal-Weight Portfolio Across Strategies

**Short.** Combine all strategy return series with equal weights and monthly rebalancing.  
This is a simple baseline for multi-strategy aggregation.


```python

def equal_weight_strategy_portfolio(strat_rets: pd.DataFrame, rebal: str = "ME") -> pd.Series:
    w = pd.DataFrame(1.0, index=strat_rets.index, columns=strat_rets.columns)
    w = normalize_rows(w)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)
    port = (strat_rets * w).sum(axis=1)
    return port

eqw_port = equal_weight_strategy_portfolio(strategies, rebal="ME")
eqw_nav  = (1 + eqw_port).cumprod()
```

## Step 8 — Risk Parity Portfolio Across Strategies

**Short.** Compute weights so that each strategy contributes an equal share of portfolio risk, using a rolling covariance estimator and monthly rebalancing.

**Risk model.**  
Let $\Sigma_t$ be the covariance matrix of strategy returns estimated on a rolling window of $W$ days.  
Portfolio variance is $\sigma_p^2 = w^\top \Sigma_t\, w$. The *marginal contribution to risk* (MCR) is
$$
\text{MCR}_i = \frac{(\Sigma_t\, w)_i}{\sigma_p}.
$$
Risk parity sets each *total* contribution $w_i \cdot \text{MCR}_i$ equal across $i$ (under $\sum_i w_i = 1$, $w_i \ge 0$).


```python

def risk_parity_weights(Sigma: np.ndarray,
                        tol: float = 1e-10,
                        max_iter: int = 10_000) -> np.ndarray:
    """
    Solve long-only risk parity weights for a positive semi-definite covariance matrix Sigma.
    Simple cyclic coordinate descent on log-barrier formulation (approximate, robust in practice).
    Returns weights summing to 1.
    """
    n = Sigma.shape[0]
    w = np.full(n, 1.0 / n)

    def port_vol(w_):
        v = w_.T @ Sigma @ w_
        return np.sqrt(max(v, 0.0))

    for _ in range(max_iter):
        w_old = w.copy()
        # Update each coordinate proportional to inverse of (Sigma w)
        Sw = Sigma @ w
        # Avoid divide-by-zero
        Sw = np.where(Sw <= 0, 1e-12, Sw)
        w = 1.0 / Sw
        # Enforce non-negativity and renormalize
        w = np.clip(w, 0, None)
        s = w.sum()
        if s == 0:
            w = np.full(n, 1.0 / n)
        else:
            w = w / s
        # Convergence
        if np.linalg.norm(w - w_old, ord=1) < tol:
            break
    return w

def combine_with_risk_parity(strat_rets: pd.DataFrame,
                             window: int = TRADING_DAYS,
                             rebal: str = "ME") -> tuple[pd.Series, pd.DataFrame]:
    """
    Build a risk-parity portfolio of strategies rebalanced monthly.
    Covariance is estimated on a rolling window of 'window' days.
    """
   
    cal_month_ends = pd.date_range(strat_rets.index.min(), strat_rets.index.max(), freq="ME")

    ix = strat_rets.index.get_indexer(cal_month_ends, method="pad")
    ix = ix[ix >= 0]
    month_ends = strat_rets.index[ix]
    weights = pd.DataFrame(0.0, index=strat_rets.index, columns=strat_rets.columns)

    for t in month_ends.unique():
        end_loc = strat_rets.index.get_loc(t)
        start_loc = max(0, end_loc - window + 1)
        block = strat_rets.iloc[start_loc:end_loc+1]

        vols = block.std()
        inv_vol = 1.0 / vols.replace(0, np.nan)
        w = (inv_vol / inv_vol.sum()).fillna(0.0)
        weights.loc[t] = w

    weights = weights.replace(0, np.nan).ffill().fillna(0.0)
    port = (strat_rets * weights).sum(axis=1)
    return port, weights

# Build risk parity portfolio across our strategy set
rp_port, rp_w = combine_with_risk_parity(strategies, window=TRADING_DAYS, rebal="ME")
rp_nav = (1 + rp_port).cumprod()
```

## Step 9 — Evaluation and Visualization

**Short.** Compare individual strategies, the equal-weight multi-strategy portfolio, and the risk-parity portfolio via NAV charts and summary metrics.


```python

# Individual strategies (NAV)
equity_strats = (1 + strategies).cumprod()

# Combined portfolios (NAV)
equity_combo = pd.DataFrame({
    "Equal-Weight (strategies)": eqw_nav,
    "Risk Parity (strategies)":  rp_nav,
}).dropna(how="any")

# Plot
ax = equity_strats.join(equity_combo, how="inner").plot(lw=1.3,
    title="Strategies and Combined Portfolios — NAV")
ax.set_ylabel("NAV")
plt.show()

# Reports
rows = []
for name in strategies.columns:
    rows.append(report_from_returns(strategies[name], name))
rows.append(report_from_returns(eqw_port, "Equal-Weight (strategies)"))
rows.append(report_from_returns(rp_port,  "Risk Parity (strategies)"))

summary = pd.concat(rows, axis=0)
summary_fmt = (
    summary.assign(**{
        "CAGR(%)": summary["CAGR"]*100,
        "Volatility(%)": summary["Volatility"]*100,
        "MaxDrawdown(%)": summary["MaxDrawdown"]*100
    })
    .drop(columns=["CAGR","Volatility","MaxDrawdown"])
    .sort_values(["Sharpe(≈)","CAGR(%)"], ascending=False)
    .round(3)
)
display(summary_fmt)
```

## Step 10 — (Optional) Sensitivity to Risk-Parity Window

**Short.** Test robustness of risk parity to the covariance lookback window $W$ (e.g., 126/252/504 days) and compare performance stability.


```python

wins = [126, 252, 504]
rows = []
rp_navs = {}

for w in wins:
    rp_p, rp_w_tmp = combine_with_risk_parity(strategies, window=w, rebal="ME")
    rp_navs[f"RP (W={w})"] = (1 + rp_p).cumprod()
    rows.append(report_from_returns(rp_p, f"RP (W={w})"))

rp_nav_panel = pd.DataFrame(rp_navs).dropna(how="any")
ax = rp_nav_panel.plot(lw=1.2, title="Risk Parity NAV — Sensitivity to Window W")
ax.set_ylabel("NAV"); plt.show()

rp_sens = pd.concat(rows, axis=0)
rp_sens_fmt = (
    rp_sens.assign(**{
        "CAGR(%)": rp_sens["CAGR"]*100,
        "Volatility(%)": rp_sens["Volatility"]*100,
        "MaxDrawdown(%)": rp_sens["MaxDrawdown"]*100
    })
    .drop(columns=["CAGR","Volatility","MaxDrawdown"])
    .round(3)
)
display(rp_sens_fmt)
```

## Step 10 — Final Discussion and Takeaways

### 1. Concept Recap
The **Risk Parity** approach allocates capital such that each asset contributes an *equal share* of total portfolio risk.  
Formally, weights are proportional to the inverse of the covariance matrix of recent returns:

$$
w_t \propto \Sigma_t^{-1}\mathbf{1}, \qquad
w_t = \frac{\Sigma_t^{-1}\mathbf{1}}{\mathbf{1}^\top \Sigma_t^{-1}\mathbf{1}}.
$$

This naturally assigns **less weight to volatile or highly correlated assets** and more weight to stable, diversifying ones.

---

### 2. Observations from Results
- The portfolio used assets `VTI`, `SPY`, `QQQ`, `AAPL`, `MSFT` — mixing broad indices and individual large-cap stocks.
- **Equal-Weight (EW)** and **Risk Parity (RP)** show broadly similar long-term growth, but RP tends to smooth the equity curve and reduce drawdowns when asset volatilities diverge.
- During periods when all assets move together (high correlations), RP loses some of its edge — risk cannot be diversified away if everything becomes “one trade.”
- When correlations and volatilities are **asymmetric** (e.g., tech stocks high-vol, ETFs stable), RP systematically shifts capital toward calmer assets, stabilizing NAV.
- The **Sharpe ratio** often improves slightly versus EW, but not always dramatically — gains depend on the heterogeneity of risks in the universe.

---

### 3. Sensitivity and Robustness
- Performance is moderately sensitive to the **lookback window** used for estimating the covariance matrix.  
  - Short windows (≈3–6 months) react faster but can over-trade.  
  - Longer windows (≈12–24 months) are more stable but slower to adapt.
- Monthly rebalancing with a **one-day execution lag** provides a good balance between responsiveness and turnover.
- Transaction costs matter less here than in Mean Reversion, since RP trades relatively infrequently.

---

### 4. Practical Interpretation
- Risk Parity shines in **multi-asset** or **diversified stock/ETF universes**, where volatilities and correlations fluctuate independently.
- When applied to **correlated equities only** (e.g., multiple US stocks or ETFs tracking the same market), it may reduce volatility but cannot meaningfully enhance returns.
- The method can serve as a **core allocation framework** into which tactical signals (momentum, valuation, ML forecasts) can later be integrated.

---

### 5. Summary Table (Interpretation)
| Metric | Meaning | Typical RP Effect |
|:--|:--|:--|
| **CAGR (%)** | Long-term annual growth | Slightly below EW when correlations high |
| **Volatility (%)** | Annualized std of daily returns | Lower than EW |
| **Sharpe (≈)** | Risk-adjusted return | Often higher |
| **Max Drawdown (%)** | Largest drop from peak | Reduced drawdown depth |

---

### 6. Key Takeaways
- **Equalized risk ≠ equal returns** — RP stabilizes exposure, not alpha.  
- Works best with **diverse, low-correlation** assets (equities, bonds, commodities, gold, etc.).  
- Avoids concentration risk that plagues equal-weight portfolios during volatility spikes.  
- In this mixed universe (`VTI`, `SPY`, `QQQ`, `AAPL`, `MSFT`), RP behaves like a *volatility smoother*: slightly lower CAGR but more stable equity curve.  
- Overall, RP provides a **robust baseline strategy** — simple, interpretable, and a strong reference for later model-driven or ML-based allocation systems.
