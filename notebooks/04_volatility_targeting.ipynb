# Volatility Targeting & Position Sizing

**Goal.** Convert raw strategy signals into *risk-adjusted* portfolios that target a stable annualized volatility level.  
This notebook:
1) loads prices,  
2) defines reusable utilities (weights/rebalancing/costs/metrics),  
3) implements rolling volatility estimation,  
4) implements volatility targeting (ex-ante) with execution lag (no look-ahead),  
5) wraps previous strategies (Equal-Weight baseline and 12M-1 Momentum) into volatility-targeted variants,  
6) compares performance versus non-targeted versions.

**Key idea.** If realized risk drifts over time, scale exposure so that
the portfolio’s *expected* annualized volatility is close to a chosen target (e.g., 12%).


## Step 1 — Imports & Config

```python

import sqlite3
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["axes.grid"] = True

DB_PATH = Path(r"C:\Users\Ruben\Desktop\Projects\InvestmentStrategies\market.db")
TABLE     = "prices"
TICKERS   = ["VTI","SPY","QQQ","AAPL","MSFT"]
START     = "2015-01-01"
END       = "2025-01-01"
REBAL     = "ME"                    # 'D' or 'M'
COST_BP   = 10                     # transaction cost per unit one-way turnover (bps)
TRADING_DAYS = 252
```

## Step 2 — Load prices

We query a wide price panel: rows = dates, columns = tickers, values = close.  
The query uses parameter binding (no string interpolation of values) and pivots the result to wide form.


```python

placeholders = ",".join([f":s{i}" for i, _ in enumerate(TICKERS)])

q = f"""
SELECT date, symbol, close
FROM {TABLE}
WHERE symbol IN ({placeholders})
  AND date BETWEEN :start AND :end
ORDER BY date ASC
"""

params = {**{f"s{i}": t for i, t in enumerate(TICKERS)},
          "start": START, "end": END}

with sqlite3.connect(DB_PATH) as con:
    df = pd.read_sql_query(q, con, params=params, parse_dates=["date"])

prices = (df.pivot(index="date", columns="symbol", values="close")
            .sort_index()
            .dropna(how="all"))

# basic hygiene
prices = prices.loc[prices.index.drop_duplicates()].ffill().dropna(how="any")

# sanity check
print("Prices shape:", prices.shape)
print("Columns:", list(prices.columns)[:10])
print("Dates:", prices.index.min().date(), "→", prices.index.max().date())
prices.head()
```

## Step 3 — Utilities: normalization, rebalancing, costs, metrics

We reuse the same mechanics across notebooks:  
- `normalize_rows` makes weights sum to 1 by date,  
- `apply_rebalancing` reduces trading frequency (daily vs month-end),  
- `portfolio_returns` computes daily P&L with turnover costs,  
- simple performance metrics for reporting.


```python

def normalize_rows(df: pd.DataFrame) -> pd.DataFrame:
    row_sum = df.sum(axis=1).replace(0, np.nan)
    return df.div(row_sum, axis=0).fillna(0.0)

def apply_rebalancing(weights: pd.DataFrame, freq: str = "D") -> pd.DataFrame:
    if freq.upper() == "D":
        return weights
    if freq.upper() == "ME":
        monthly = weights.resample("ME").last()
        return monthly.reindex(weights.index).ffill().fillna(0.0)
    raise ValueError("freq must be 'D' or 'ME'")

def portfolio_returns(returns: pd.DataFrame,
                      weights: pd.DataFrame,
                      cost_bp: int = 0) -> pd.Series:
    returns = returns.reindex(weights.index).fillna(0.0)
    port = (returns * weights).sum(axis=1)
    if cost_bp and cost_bp > 0:
        dw = (weights - weights.shift(1)).abs().sum(axis=1) * 0.5
        port = port - dw * (cost_bp * 1e-4)
    return port

def cagr_from_nav(nav: pd.Series, trading_days: int = TRADING_DAYS) -> float:
    nav = nav.dropna()
    if nav.empty:
        return 0.0
    years = len(nav) / trading_days
    return float((nav.iloc[-1] / nav.iloc[0])**(1/years) - 1)

def sharpe(returns: pd.Series, trading_days: int = TRADING_DAYS) -> float:
    r = returns.dropna()
    if r.empty:
        return 0.0
    mu = r.mean() * trading_days
    sd = r.std(ddof=0) * np.sqrt(trading_days)
    return float(mu / sd) if sd > 0 else 0.0

def max_drawdown(nav: pd.Series) -> float:
    roll_max = nav.cummax()
    dd = nav / roll_max - 1.0
    return float(dd.min())

def report_from_returns(r: pd.Series, name: str) -> pd.DataFrame:
    nav = (1 + r).cumprod()
    out = pd.DataFrame({
        "Strategy": [name],
        "CAGR": [cagr_from_nav(nav)],
        "Volatility": [r.std(ddof=0) * np.sqrt(TRADING_DAYS)],
        "Sharpe(≈)": [sharpe(r)],
        "MaxDrawdown": [max_drawdown(nav)],
    }).set_index("Strategy")
    return out.round(6)
```

## Step 4 — Baselines to be risk-managed

We will wrap two existing strategies:
1) **Equal-Weight baseline** (rebalance at chosen frequency),  
2) **12M Momentum with last-month skip (12-1)** at month-end.

Signals are shifted by one day to avoid look-ahead before converting to weights.


```python

def bh_equal(prices: pd.DataFrame, rebal: str = "ME", cost_bp: int = 0) -> pd.Series:
    rets = prices.pct_change()
    w = pd.DataFrame(1.0, index=prices.index, columns=prices.columns)
    w = normalize_rows(w)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)  # execute next day
    return portfolio_returns(rets, w, cost_bp=cost_bp)

def momentum_12m(prices: pd.DataFrame,
                 lookback: int = TRADING_DAYS,
                 skip_last_month: bool = True,
                 top_k: int | None = None,
                 rebal: str = "ME",
                 cost_bp: int = 0) -> pd.Series:
    rets = prices.pct_change()
    shift_curr = 21 if skip_last_month else 0
    mom = prices.shift(shift_curr) / prices.shift(lookback) - 1.0
    mom_m = mom.resample("ME").last()

    if top_k is not None and top_k > 0:
        ranks = mom_m.rank(axis=1, method="first", ascending=False)
        sel = (ranks <= top_k).astype(float)
    else:
        sel = (mom_m > 0).astype(float)

    w_m = normalize_rows(sel)
    w = w_m.reindex(prices.index).ffill().fillna(0.0)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)  # execute next day
    return portfolio_returns(rets, w, cost_bp=cost_bp)
```

## Step 5 — Rolling volatility estimation

We estimate ex-ante volatility from a rolling window of past daily returns.

- Let $r_t$ be daily simple returns.  
- Rolling standard deviation over $W$ days: $\hat{\sigma}_t = \text{std}(r_{t-W+1:t})$.  
- Annualized estimate: $\hat{\sigma}^{(ann)}_t = \hat{\sigma}_t \sqrt{252}$.

We will apply this to *portfolio* returns implied by raw (non-targeted) weights to compute a scaling factor.


```python

def rolling_annualized_vol(returns: pd.Series, window: int = 60, trading_days: int = TRADING_DAYS) -> pd.Series:
    vol = returns.rolling(window, min_periods=window).std(ddof=0)
    return vol * np.sqrt(trading_days)
```

## Step 6 — Volatility Targeting (Position Scaling)

Volatility targeting does **not** try to change market volatility itself — it adjusts how much of our portfolio participates in that volatility.  
When markets are calm, we slightly increase exposure (invest a bit more, or use mild leverage).  
When markets are turbulent, we reduce exposure (hold smaller positions, or keep some capital in cash).  
This way, our portfolio's *own* volatility — its “amplitude of motion” — stays near a constant target (for example, $12\%$ per year).

In other words, we are not modifying returns directly — we are **rescaling portfolio weights** so that the resulting portfolio would *behave as if* its volatility were always close to $12\%$, even though market volatility itself changes every day.

We achieve this by estimating recent volatility, comparing it to the target, and scaling all weights proportionally:
- If estimated volatility is lower than target → scale $>$ 1 → take slightly larger positions.
- If estimated volatility is higher than target → scale $<$ 1 → take smaller positions.
- `L_{max}` puts a limit on how much we can increase exposure (to avoid extreme leverage).

---

This step applies **volatility targeting**, a common technique in quantitative portfolio management that keeps portfolio risk stable through time.

---

### Concept

Raw strategies (e.g., momentum or moving-average crossovers) often experience large fluctuations in volatility.  
To maintain a consistent risk level, we dynamically scale portfolio exposure so that the *expected volatility* remains near a chosen target, for example $12\%$ annualized.

Formally, we adjust the portfolio’s daily weights and returns based on the estimated rolling volatility.

---

### Mathematical Framework

1. **Estimate recent volatility**

   $$\hat{\sigma}_t^{(ann)} = \text{std}(r_{t-W+1:t}) \times \sqrt{252}$$  

   where  
   - $r_t$ — daily returns,  
   - $W$ — estimation window (e.g., 60 days),  
   - $\hat{\sigma}_t^{(ann)}$ — annualized volatility estimate.

2. **Compute the scaling factor**

   $$\text{scale}_t = \min\left(\frac{\sigma^*}{\hat{\sigma}_{t-1}^{(ann)}}, L_{\max}\right)$$  

   where  
   - $\sigma^*$ — target volatility (e.g., $0.12$),  
   - $L_{\max}$ — maximum leverage cap (e.g., $1.5$),  
   - the one-day lag $(t-1)$ avoids look-ahead bias.

3. **Scale weights and returns**

   $$
   \begin{aligned}
   w_t^{(VT)} &= w_t^{(raw)} \times \text{scale}_t, \\
   r_t^{(VT)} &= r_t^{(raw)} \times \text{scale}_t.
   \end{aligned}
   $$

   The resulting volatility-targeted portfolio has a smoother risk profile and more consistent return-to-risk ratio.

---

### Implementation Notes

- `rolling_annualized_vol()` estimates the realized volatility using a moving window.  
- `scale` is capped and shifted by one day to avoid look-ahead bias.  
- `raw_weights` and `raw_returns` are multiplied by `scale` to produce `vt_weights` and `vt_returns`.

---

### Intuition

When recent volatility increases, the strategy automatically reduces its exposure;  
when volatility falls, it increases exposure — keeping the *expected risk level* roughly constant.  
This approach helps to stabilize performance and improve comparability across strategies with different inherent volatilities.


```python

def apply_vol_targeting(raw_returns: pd.Series,
                        raw_weights: pd.DataFrame,
                        target_vol: float = 0.12,
                        est_window: int = 60,
                        max_leverage: float = 1.5) -> tuple[pd.Series, pd.DataFrame]:
    """
    Inputs:
      raw_returns: daily portfolio returns from a raw strategy (without vol targeting)
      raw_weights: daily portfolio weights from the same strategy (pre-cost, pre-scale)
    Outputs:
      vt_returns: daily returns after volatility scaling (no costs applied here)
      vt_weights: scaled weights (capped), with one-day lag
    """
    # ex-ante vol estimate based on *past* realized returns
    est_vol = rolling_annualized_vol(raw_returns, window=est_window)

    # scale(t) uses yesterday's estimate
    scale = (target_vol / est_vol).clip(lower=0.0, upper=max_leverage).shift(1).fillna(0.0)

    # scale weights and keep orientation; optional renormalization is not needed for pure scaling
    vt_weights = (raw_weights.mul(scale, axis=0)).fillna(0.0)

    # the corresponding returns are just scaled raw returns (pre-cost here)
    vt_returns = (raw_returns * scale).fillna(0.0)

    return vt_returns, vt_weights
```

## Step 7 — Get raw weights to feed the scaler

To apply targeting properly, we need the *raw* weights used to produce raw returns.  
Below we provide helper functions that return both: raw returns **and** the aligned raw weights.


```python

def bh_equal_rw(prices: pd.DataFrame, rebal: str = "ME") -> tuple[pd.Series, pd.DataFrame]:
    rets = prices.pct_change()
    w = pd.DataFrame(1.0, index=prices.index, columns=prices.columns)
    w = normalize_rows(w)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)  # execute next day
    r = (rets * w).sum(axis=1)  # NO costs here; costs will be applied later if desired
    return r, w

def momentum_12m_rw(prices: pd.DataFrame,
                    lookback: int = TRADING_DAYS,
                    skip_last_month: bool = True,
                    top_k: int | None = None,
                    rebal: str = "ME") -> tuple[pd.Series, pd.DataFrame]:
    rets = prices.pct_change()
    shift_curr = 21 if skip_last_month else 0
    mom = prices.shift(shift_curr) / prices.shift(lookback) - 1.0
    mom_m = mom.resample("ME").last()

    if top_k is not None and top_k > 0:
        ranks = mom_m.rank(axis=1, method="first", ascending=False)
        sel = (ranks <= top_k).astype(float)
    else:
        sel = (mom_m > 0).astype(float)

    w_m = normalize_rows(sel)
    w = w_m.reindex(prices.index).ffill().fillna(0.0)
    w = apply_rebalancing(w, rebal)
    w = w.shift(1).fillna(0.0)

    r = (rets * w).sum(axis=1)  # raw, no costs
    return r, w
```

## Step 8 — Build targeted variants and compute net returns

We compute:
- raw returns and raw weights,  
- volatility-targeted returns/weights,  
- apply transaction costs with the same cost model as before,  
- produce final net returns and NAV.


```python

def net_from_weights(returns: pd.DataFrame, weights: pd.DataFrame, cost_bp: int) -> pd.Series:
    return portfolio_returns(returns, weights, cost_bp=cost_bp)

# Raw baselines (non-targeted)
bh_r_raw,  bh_w_raw  = bh_equal_rw(prices, rebal=REBAL)
mom_r_raw, mom_w_raw = momentum_12m_rw(prices, lookback=TRADING_DAYS, skip_last_month=True,
                                       top_k=None, rebal="ME")

# Volatility-targeted variants (target 12%)
VTARGET = 0.12
EST_W   = 60
LEV_MAX = 1.5

bh_r_vt,  bh_w_vt  = apply_vol_targeting(bh_r_raw,  bh_w_raw,  target_vol=VTARGET, est_window=EST_W, max_leverage=LEV_MAX)
mom_r_vt, mom_w_vt = apply_vol_targeting(mom_r_raw, mom_w_raw, target_vol=VTARGET, est_window=EST_W, max_leverage=LEV_MAX)

# Convert scaled weights into net returns with costs
rets = prices.pct_change()

bh_net_raw  = net_from_weights(rets, bh_w_raw,  cost_bp=COST_BP)
mom_net_raw = net_from_weights(rets, mom_w_raw, cost_bp=COST_BP)

bh_net_vt   = net_from_weights(rets, bh_w_vt,   cost_bp=COST_BP)
mom_net_vt  = net_from_weights(rets, mom_w_vt,  cost_bp=COST_BP)
```

## Step 9 — Equity curves and summary

We compare non-targeted versus targeted versions on a common time axis.


```python

equity = pd.DataFrame({
    "B&H Equal (raw)": (1 + bh_net_raw).cumprod(),
    "B&H Equal (vol-targeted)": (1 + bh_net_vt).cumprod(),
    "12M-1 (raw)": (1 + mom_net_raw).cumprod(),
    "12M-1 (vol-targeted)": (1 + mom_net_vt).cumprod(),
}).dropna(how="any")

ax = equity.plot(lw=1.3, title=f"Equity Curves — Volatility Targeting (target={int(100*VTARGET)}% p.a., window={EST_W}d, Lmax={LEV_MAX}x)")
ax.set_ylabel("NAV")
plt.show()

summary = pd.concat([
    report_from_returns(bh_net_raw,  "B&H Equal (raw)"),
    report_from_returns(bh_net_vt,   "B&H Equal (vol-targeted)"),
    report_from_returns(mom_net_raw, "12M-1 (raw)"),
    report_from_returns(mom_net_vt,  "12M-1 (vol-targeted)"),
], axis=0)

summary_fmt = (
    summary.assign(**{
        "CAGR(%)": summary["CAGR"]*100,
        "Volatility(%)": summary["Volatility"]*100,
        "MaxDrawdown(%)": summary["MaxDrawdown"]*100
    })
    .drop(columns=["CAGR","Volatility","MaxDrawdown"])
    .sort_values(["Sharpe(≈)","CAGR(%)"], ascending=False)
    .round(3)
)

display(summary_fmt)
```

## Step 10 — Sensitivity analysis (optional)

Assess robustness across different targets and windows.


```python

targets  = [0.10, 0.12, 0.15]   # target annual vol (10%, 12%, 15%)
windows  = [40, 60, 90]         # rolling window for vol estimate (days)

rows = []
for tv in targets:
    for w in windows:
        # 1) Raw strategy: 12M momentum with last-month skip, month-end rebalance
        r_raw, w_raw = momentum_12m_rw(
            prices,
            lookback=TRADING_DAYS,      # ~252
            skip_last_month=True,       # 12-1 momentum
            top_k=None,                 # long-only, include all with positive score
            rebal="ME",                 # month-end
        )

        # 2) Apply volatility targeting (returns + scaled weights)
        r_vt, w_vt = apply_vol_targeting(
            r_raw,
            w_raw,
            target_vol=tv,
            est_window=w,
            max_leverage=LEV_MAX
        )

        # 3) Convert scaled weights back to net daily returns with costs (same model as before)
        net = net_from_weights(prices.pct_change(), w_vt, cost_bp=COST_BP)

        # 4) Summarize metrics; store Target/Window so we can pivot for heatmaps
        rep = report_from_returns(net, name=f"12M-1 VT (σ*={int(tv*100)}%, W={w})")
        rep = rep.assign(Target=int(tv*100), Window=w)  # add for pivoting
        rows.append(rep)

# Collect the sensitivity table
sens = pd.concat(rows, axis=0)

# Nicely formatted table (percentages) sorted by Sharpe, then CAGR
sens_fmt = (
    sens.assign(**{
        "CAGR(%)":       sens["CAGR"]*100,
        "Volatility(%)": sens["Volatility"]*100,
        "MaxDrawdown(%)": sens["MaxDrawdown"]*100,
    })
    .drop(columns=["CAGR", "Volatility", "MaxDrawdown"])
    .sort_values(["Sharpe(≈)", "CAGR(%)"], ascending=[False, False])
    .round(3)
)

display(sens_fmt)

# Heatmaps
def _heatmap_from_sens(df, value_col, title, cmap="viridis"):
    """Build a heatmap from `sens` for a single metric."""
    # Build pivot (rows: Target %, cols: Window days)
    pivot = (
        df.pivot_table(index="Target", columns="Window", values=value_col)
        .sort_index(axis=0)  # Target axis
        .sort_index(axis=1)  # Window axis
    )

    fig, ax = plt.subplots(figsize=(8, 4.2))
    im = ax.imshow(pivot.values, aspect="auto", cmap=cmap)

    # Axis ticks & labels
    ax.set_yticks(range(len(pivot.index)))
    ax.set_yticklabels([f"{t}%" for t in pivot.index])
    ax.set_xticks(range(len(pivot.columns)))
    ax.set_xticklabels(pivot.columns)
    ax.set_xlabel("Estimation window W (days)")
    ax.set_ylabel("Target volatility σ* (%)")
    ax.set_title(title)

    # Annotate cells
    for i in range(pivot.shape[0]):
        for j in range(pivot.shape[1]):
            val = pivot.values[i, j]
            if pd.notna(val):
                ax.text(j, i, f"{val:.2f}", ha="center", va="center", fontsize=9)

    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()

# Heatmap for Sharpe
_heatmap_from_sens(sens.assign(Sharpe=sens["Sharpe(≈)"]),
                   value_col="Sharpe",
                   title="Heatmap — Sharpe by Target σ* and Window W")

# Heatmap for CAGR (%)
_heatmap_from_sens(sens.assign(CAGR_pct=sens["CAGR"]*100),
                   value_col="CAGR_pct",
                   title="Heatmap — CAGR (%) by Target σ* and Window W")
```

## Step 10 — Sensitivity Analysis (Target Volatility × Estimation Window)

This step stress-tests volatility targeting for the **12M momentum (12-1)** strategy across a grid of:
- **Target annual volatility** levels: $10\%$, $12\%$, $15\%$; and
- **Estimation windows** for the rolling volatility used by the scaler: $W \in \{40, 60, 90\}$ days.

For each (target, window) pair we:
1. Compute **raw** strategy returns and weights (12-1 momentum, month-end rebalance).
2. Apply **volatility targeting** to obtain scaled daily weights/returns using the windowed vol estimate.
3. Convert scaled weights back to **net daily returns** (same transaction-cost model as earlier).
4. Summarize **Sharpe**, **CAGR**, **Volatility**, and **Max Drawdown**.

### Why this matters
Volatility targeting adds an extra control knob (risk level and estimation horizon). A robust setting should not be overly sensitive to small parameter changes; performance should be **stable** across nearby choices of $σ^\*$ and $W$.

### Reading the heatmaps
- **Sharpe heatmap** highlights where risk-adjusted performance peaks.
- **CAGR heatmap** shows pure growth response to the scaling choices.
- Rows correspond to **target vol** $σ^\*$ (in %), columns to **window** $W$ (days). 
- Each cell is annotated with the metric value for that configuration.

### Typical patterns to look for
- **Too short** a window (very small $W$) can make the scaler overreact to noise, harming Sharpe.
- **Too long** a window (very large $W$) adjusts too slowly, leaving risk high in shocks.
- A **mid-range** window (e.g., 60 days) often balances responsiveness and stability.
- Raising the **target** $σ^\*$ increases average exposure (and drawdowns) and may raise CAGR but not necessarily Sharpe.

### Takeaway
Use this grid to pick a **stable** configuration rather than the single best in-sample cell. If multiple adjacent cells around $(σ^\*, W)$ deliver comparable Sharpe and CAGR, that region is a safer choice for live usage than a lone outlier.
